```python
def AND_gate(x1, x2):
    w1=0.5
    w2=0.5
    b=-0.7
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
print(AND_gate(1,1))
```

```python
def NAND_gate(x1, x2):
    w1=-0.5
    w2=-0.5
    b=0.7
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
print(AND_gate(0,0))
```

```python
def OR_gate(x1, x2):
    w1=0.6
    w2=0.6
    b=-0.5
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
print(OR_gate(0.5,0.5))
```

```python
def XOR_gate(x1, x2):
    nx = NAND_gate(x1, x2)
    ox = OR_gate(x1, x2)
    result = AND_gate(nx,ox)
    return result;
```

```python
XOR_gate(0,1)
```
